<h1>Test-Training-Split for V1</h1>

```python
df1 = pd.read_csv('v1_data_dog_prepared.csv')
df1.head()
```

```python
X= df1['size_dog']
y= df1['eye_circularity']

# using the train test split function
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, shuffle=True)

# printing out train and test sets
print('X_train : ')
print(X_train.head())
print(X_train.shape)
print('')
print('X_test : ')
print(X_test.head())
print(X_test.shape)
print('')
print('y_train : ')
print(y_train.head())
print(y_train.shape)
print('')
print('y_test : ')
print(y_test.head())
print(y_test.shape)
```
